```python
class Produto:
    def __init__(self, nome, preco, quantidade):
        self.nome = nome
        self.preco = preco
        self.quantidade = quantidade

    def exibir_detalhes(self):
        print("Nome:", self.nome)
        print("Preço:", self.preco)
        print("Quantidade:", self.quantidade)


class Cliente:
    def __init__(self, nome, cpf, endereco):
        self.nome = nome
        self.cpf = cpf
        self.endereco = endereco

    def exibir_detalhes(self):
        print("Nome do cliente:", self.nome)
        print("CPF do cliente:", self.cpf)
        print("Endereço do cliente:", self.endereco)


def cadastrar_produto():
    nome = input("Digite o nome do produto: ")
    preco = float(input("Digite o preço do produto: "))
    quantidade = int(input("Digite a quantidade do produto: "))
    return Produto(nome, preco, quantidade)


def cadastrar_cliente():
    nome = input("Digite o nome do cliente: ")
    cpf = input("Digite o CPF do cliente: ")
    endereco = input("Digite o endereço do cliente: ")
    return Cliente(nome, cpf, endereco)


def realizar_venda(produtos, cliente):
    itens_vendidos = []
    total_a_pagar = 0

    print("\nProdutos disponíveis:")
    for i, produto in enumerate(produtos):
        print(f"{i + 1}: {produto.nome} - Preço: R${produto.preco} - Quantidade disponível: {produto.quantidade}")

    while True:
        escolha = int(input("Escolha o número do produto que deseja comprar (0 para finalizar): ")) - 1

        if escolha == -1:
            break

        if escolha < 0 or escolha >= len(produtos):
            print("Escolha inválida!")
            continue

        quantidade_comprar = int(input("Digite a quantidade que deseja comprar: "))
        produto_escolhido = produtos[escolha]

        if produto_escolhido.quantidade < quantidade_comprar:
            print("Quantidade insuficiente em estoque!")
            continue

        produto_escolhido.quantidade -= quantidade_comprar
        itens_vendidos.append((produto_escolhido, quantidade_comprar))
        total_a_pagar += produto_escolhido.preco * quantidade_comprar

    print("\nVenda realizada com sucesso!")
    print("Itens comprados:")
    for produto, quantidade in itens_vendidos:
        print(f"Produto: {produto.nome} - Preço unitário: R${produto.preco} - Quantidade: {quantidade}")
    print(f"Total a pagar: R${total_a_pagar}")
    print("Cliente:")
    cliente.exibir_detalhes()


def main():
    produtos = []
    clientes = []

    while True:
        print("\n1. Cadastrar Produto")
        print("2. Cadastrar Cliente")
        print("3. Realizar Venda")
        print("4. Sair")

        opcao = int(input("Escolha uma opção: "))

        if opcao == 1:
            produto = cadastrar_produto()
            produtos.append(produto)
            print("Produto cadastrado com sucesso!")

        elif opcao == 2:
            cliente = cadastrar_cliente()
            clientes.append(cliente)
            print("Cliente cadastrado com sucesso!")

        elif opcao == 3:
            if not produtos:
                print("Não há produtos cadastrados!")
                continue

            if not clientes:
                print("Não há clientes cadastrados!")
                continue

            cliente = clientes[0]  # Para simplificar, estamos pegando sempre o primeiro cliente cadastrado
            realizar_venda(produtos, cliente)

        elif opcao == 4:
            print("Saindo...")
            break

        else:
            print("Opção inválida!")


if __name__ == "__main__":
    main()
]
```
